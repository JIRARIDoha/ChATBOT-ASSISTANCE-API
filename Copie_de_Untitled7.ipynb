```python
import pandas as pd
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

# 1. Création d'un mini-dataset de secours pour générer les fichiers de l'IA
data = {
    'texte_darija': [
        'بغيت دوليبران عافاك', 'فيا السخانة و راسي كيحرقني', 'فين جات أقرب فرماسي',
        'خاصني جوج باكيات سيرو', 'عتقوني راجل غيب دابا عافاك', 'بشحال كيدير هاد الدوا'
    ],
    'commande': [1, 0, 0, 1, 0, 0],
    'symptome': [0, 1, 0, 0, 1, 0],
    'information': [0, 0, 1, 0, 0, 1],
    'quantite': [0, 0, 0, 1, 0, 0],
    'urgence': [0, 0, 0, 0, 1, 0]
}
df = pd.DataFrame(data)

# 2. Entraînement ultra-rapide
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['texte_darija'])
y = df[['commande', 'symptome', 'information', 'quantite', 'urgence']]

model = OneVsRestClassifier(LogisticRegression())
model.fit(X, y)

# 3. Sauvegarde des fichiers .pkl indispensables
joblib.dump(model, 'classifieur_medical_darija.pkl')
joblib.dump(vectorizer, 'vectorizer_darija.pkl')
print("🎉 Les fichiers modèles .pkl ont été créés avec succès !")
```

```python
%%writefile app.py
import streamlit as st
import joblib
import re

# 1. Chargement direct à chaque analyse (Pas de cache blockant !)
def load_model():
    try:
        model = joblib.load('classifieur_medical_darija.pkl')
        vectorizer = joblib.load('vectorizer_darija.pkl')
    except:
        model, vectorizer = None, None
    return model, vectorizer

# 2. Extraction des informations (NER)
def extraire_entites(texte):
    liste_medicaments = [
        'دوليبران', 'doliprane', 'أنيبيوتيك', 'antibiotique', 'سيرو', 'sirop',
        'اسبرين', 'aspirine', 'بانادول', 'panadol', 'فيتامين', 'vitamin', 'دوا'
    ]
    meds_trouves = [med for med in liste_medicaments if med in texte.lower()]

    regex_quantite = r'(\b\d+\b|\bجوج\b|\bتلاتة\b|\bواحد\b|\bشوية\b)\s*(باكية|بكية|قرعة|حبة|غرام|mg|g|بواط|boite|comprimé)?'
    quantites_trouvees = re.findall(regex_quantite, texte.lower())
    quants_nettoyees = [" ".join(q).strip() for q in quantites_trouvees if q[0]]

    return meds_trouves, quants_nettoyees

# 3. Interface Graphique
st.set_page_config(page_title="Chatbot Médical UM5", page_icon="🩺", layout="centered")

st.title("🩺 Assistant Médical Intelligent (Darija/Arabic)")
st.write("Projet Master IT 2026 - Analyse d'intentions et Extraction d'informations (NER).")
st.markdown("---")

user_input = st.text_input("Saisissez le message du patient :", "عافاك بغيت جوج باكيات ديال دوليبران دابا")

if st.button("Analyser la demande"):
    if user_input.strip():
        st.subheader("🎯 1. Classification de l'Intention")

        # On force la recharge ici
        model, vectorizer = load_model()

        if model and vectorizer:
            vec_text = vectorizer.transform([user_input])
            pred = model.predict(vec_text)
            classes = ['commande', 'information', 'symptome', 'quantite', 'urgence']
            intentions = [classes[idx].upper() for idx, nom_classe in enumerate(classes) if pred[0][idx] == 1]

            if intentions:
                st.success(f"Intentions détectées : **{', '.join(intentions)}**")
            else:
                st.info("✅ INTENTION : AUTRE / INFORMATION")
        else:
            st.warning("⚠️ Mode démonstration activé (Modèles .pkl non détectés).")

        st.subheader("📦 2. Extraction d'Informations Clés (NER)")
        medicaments, quantites = extraire_entites(user_input)

        col1, col2 = st.columns(2)
        with col1:
            st.markdown("**💊 Médicament(s) trouvé(s) :**")
            if medicaments:
                for m in medicaments: st.markdown(f"- `{m}`")
            else:
                st.caption("Aucun médicament détecté")

        with col2:
            st.markdown("**🔢 Quantité(s) trouvée(s) :**")
            if quantites:
                for q in quantites: st.markdown(f"- `{q}`")
            else:
                st.caption("Aucune quantité détectée")
```
