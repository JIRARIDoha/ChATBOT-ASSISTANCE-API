```python
# 1. Détection automatique de la colonne qui contient le Français
colonne_fr = None
for col in df_hf.columns:
    if 'fr' in col.lower() or 'french' in col.lower():
        colonne_fr = col
        break

if not colonne_fr:
    colonne_fr = df_hf.columns[0]

print(f"🔍 Colonne sélectionnée pour l'analyse en Français : {colonne_fr}")

# 2. Création des listes de mots-clés médicaux
mots_commande = ['acheter', 'commander', 'ordonnance', 'achat', 'pharmacie', 'boite', 'sirop', 'médicament']
mots_symptome = ['douleur', 'mal', 'fièvre', 'toux', 'vomir', 'symptôme', 'malade', 'brûlure', 'fatigue']
mots_urgence = ['urgent', 'ambulance', 'hémorragie', 'inconscient', 'étouffe', 'crise', 'vite', 'grave']

# 3. Nettoyage et application de la logique
df_clean = pd.DataFrame()
df_clean['texte'] = df_hf[colonne_fr].astype(str)

def determiner_classe(texte):
    txt = texte.lower()
    if any(m in txt for m in mots_urgence):
        return 3  # URGENCE
    elif any(m in txt for m in mots_commande):
        return 0  # COMMANDE
    elif any(m in txt for m in mots_symptome):
        return 1  # SYMPTÔME
    else:
        return 2  # INFORMATION GÉNÉRALE (Classe par défaut)

df_clean['label'] = df_clean['texte'].apply(determiner_classe)
print("✅ Génération des classes et étiquetage terminés avec succès !")
```

```python
dictionnaire_classes = {
    0: "COMMANDE / ACHAT",
    1: "SYMPTÔME / MALADIE",
    2: "INFORMATION GÉNÉRALE",
    3: "URGENCE"
}

print("="*60)
print("📊 DISTRIBUTION DES CLASSES OBTENUES")
print("="*60)

comptage = df_clean['label'].value_counts()
for idx, total in comptage.items():
    nom_reel = dictionnaire_classes.get(idx, f"Classe {idx}")
    print(f"🔹 {nom_reel} : {total} phrases ({ (total / len(df_clean)) * 100:.2f}%)")

print("\n" + "-"*50)
print("👀 APERÇU DE VOS TEXTES EN FRANÇAIS :")
print("-"*50)

for id_classe, nom_reel in dictionnaire_classes.items():
    extraits = df_clean[df_clean['label'] == id_classe]['texte'].head(2).tolist()
    print(f"\n👉 {nom_reel} :")
    if extraits:
        for i, ex in enumerate(extraits, 1):
            print(f"   [{i}] {ex[:120]}...")
    else:
        print("   ⚠️ Aucune phrase dans cette catégorie.")
```

```python
import torch
from torch.utils.data import Dataset
from transformers import AutoTokenizer

nom_modele = "distilbert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(nom_modele)

# Encodage de la colonne 'texte'
encodings = tokenizer(list(df_clean['texte']), truncation=True, padding=True, max_length=128)

# Création du Dataset PyTorch standardisé
class MedQADataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
    def __len__(self):
        return len(self.labels)

dataset_pour_ia = MedQADataset(encodings, list(df_clean['label']))
print("✅ Les données sont tokenisées et prêtes pour l'entraînement !")
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

print("🧱 Configuration des paramètres de vitesse maximale...")

# Configuration des arguments d'entraînement optimisés "Lightweight"
arguments_ultra_rapides = TrainingArguments(
    output_dir='./resultats_french',
    num_train_epochs=2,                  # 2 époques suffisent largement pour la validation du Master
    per_device_train_batch_size=32,      # On monte à 32 : le GPU traite de très gros paquets d'un coup
    fp16=True,                           # Activer le calcul accéléré sur GPU T4 (Vitesse x3)
    learning_rate=5e-5,
    weight_decay=0.01,
    logging_steps=5,
    report_to="none",                    # Désactive les outils de tracking en ligne qui ralentissent
    save_strategy="no"                   # Zéro écriture sur le disque pendant l'entraînement pour aller vite
)

# Chargement du modèle de base
model = AutoModelForSequenceClassification.from_pretrained(nom_modele, num_labels=4)
print("✅ Arguments de vitesse et architecture du modèle prêts !")
```

```python
# Initialisation du Trainer avec les paramètres accélérés
trainer = Trainer(
    model=model,
    args=arguments_ultra_rapides,
    train_dataset=dataset_pour_ia,
)

print("🚀 Lancement du Fine-Tuning Flash...")
trainer.train()
print("⚡ Entraînement terminé en un temps record !")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 1. Chargement de votre modèle sauvegardé
chemin_modele = "./modele_distilbert_français"
print("🔄 Chargement de votre IA personnalisée...")
tokenizer = AutoTokenizer.from_pretrained(chemin_modele)
model = AutoModelForSequenceClassification.from_pretrained(chemin_modele)
model.eval()

# 2. Vos phrases de test (Vous pouvez les modifier pour tester !)
phrases_tests = [
    "Je ressens une terrible douleur à la poitrine et j'ai du mal à respirer, venez vite.",
    "Bonjour, je voudrais acheter une boîte de paracétamol et du sirop pour la toux s'il vous plaît.",
    "Quels sont les effets secondaires du vaccin contre la grippe chez les enfants ?"
]

# Correspondance des classes
liste_classes = ['COMMANDE / ACHAT', 'SYMPTÔME / MALADIE', 'INFORMATION GÉNÉRALE', 'URGENCE']

print("\n🎯 Lancement des prédictions de test :")
print("="*60)

# 3. Calcul des prédictions
for i, phrase in enumerate(phrases_tests, 1):
    # Tokenisation de la phrase
    inputs = tokenizer(phrase, return_tensors="pt", truncation=True, padding=True, max_length=128)

    with torch.no_grad():
        outputs = model(**inputs)

    # Calcul des probabilités
    scores = torch.softmax(outputs.logits, dim=1).flatten().tolist()
    id_classe_predite = torch.argmax(outputs.logits, dim=1).item()

    # Affichage du résultat
    print(f"\n📝 Test n°{i} : '{phrase}'")
    print(f"🤖 Résultat IA : {liste_classes[id_classe_predite]} (Confiance : {scores[id_classe_predite]*100:.2f}%)")
    print("-" * 40)
```
